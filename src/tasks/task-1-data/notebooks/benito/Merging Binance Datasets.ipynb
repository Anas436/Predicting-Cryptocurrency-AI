# Import Data

## BTC

```python
# Importing one csv file
pd.read_csv("/content/drive/MyDrive/Colab Notebooks/01 Omdena/Bahrain - Crypto/Project_Data/binance/BTCUSDT/BTCUSDT-1h-2019-09.csv", header=None)
```

According to binance these are the columns from the csv files:

- open_time: Kline Open time in unix time format

- open: Open Price

- high: High Price

- low: Low Price

- close: Close Price

- volume: Volume

- close_time: Kline Close time in unix time format

- quote_volume: Quote Asset Volume

- count: Number of Trades

- taker_buy_volume: Taker buy quote asset volume during this period

- taker_buy_quote_volume: Taker buy base asset volume during this period

- ignore : Ignore(youcan drop this feature)

```python
# Create a list of the files

folder_path_BTC = '/content/drive/MyDrive/Colab Notebooks/01 Omdena/Bahrain - Crypto/Project_Data/binance/BTCUSDT'

file_list_BTC = os.listdir(folder_path_BTC)
```

```python
# Merging all files

dfs_BTC = []

for i in file_list_BTC:
  dfs_BTC.append(pd.read_csv(f"/content/drive/MyDrive/Colab Notebooks/01 Omdena/Bahrain - Crypto/Project_Data/binance/BTCUSDT/{i}", header=None))

# Concatenate all data into one DataFrame
data_BTC = pd.concat(dfs_BTC, ignore_index=True)

data_BTC.head()
```

```python
data_BTC.shape
```

```python
# Renaming the columns

data_BTC.columns = [
    'open_time',
    'open',
    'high',
    'low',
    'close',
    'volume',
    'close_time',
    'quote_volume',
    'count',
    'taker_buy_volume',
    'taker_buy_quote_volume',
    'ignore'
]
```

```python
# Removing ignore column

data_BTC.drop(labels="ignore", axis=1, inplace=True)
```

```python
# Changing to datetime

data_BTC["open_time"] = pd.to_datetime(data_BTC["open_time"], origin='unix',unit='ms')
data_BTC["close_time"] = pd.to_datetime(data_BTC["close_time"], origin='unix',unit='ms')
```

```python
data_BTC.head()
```

```python
# Sorting Ascending

data_BTC.sort_values(by = "open_time", ascending=True, inplace=True)
```

```python
data_BTC.head()
```

```python
data_BTC.tail()
```

```python
data_BTC.info()
```

```python
# Saving BTC merged file

data_BTC.to_csv("/content/drive/MyDrive/Colab Notebooks/01 Omdena/Bahrain - Crypto/Project_Data/binance/BTCUSDT_merged.csv", index=False)
```
